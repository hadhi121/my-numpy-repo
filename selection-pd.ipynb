```python
import pandas as pd

df = pd.read_csv('pokemon.csv',index_col='Name')
print(df.to_string())#it will show all the table
```

```python
#selecting multiple col
df[['Name','Height','Weight']]
```

```python
#print(df.loc['Pikachu'])
print(df.loc['Charizard',['Height','Weight']])#can access the only detail we want
```

```python
#selecting a range of row(not one row)
print(df.loc['Charizard':'Blastoise',['Height','Weight']])
```

```python
print(df.iloc[0:3 , 1:])
```

```python
pokemon = input('enter a pokemon name: ')
try:
    print(df.loc[pokemon])
except KeyError:
    print(f'{pokemon} does not exist.')
```
